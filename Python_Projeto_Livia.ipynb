O UCI Machine Learning Repository é um repositório de conjuntos de dados de aprendizado de máquina que oferece uma ampla variedade de conjuntos de dados, incluindo conjuntos de dados de classificação, regressão, agrupamento, associação, séries temporais e muitos outros. É amplamente utilizado por pesquisadores de todo o mundo para testar algoritmos de aprendizado de máquina, criar modelos e resolver problemas do mundo real.

Um dos conjuntos de dados disponíveis no UCI Machine Learning Repository é o "Image Segmentation Data Set", que contém informações extraídas de imagens digitais de seções transversais de tecidos fotográficos humanos. Este conjunto de dados contém 210 amostras, com 10 amostras por classe, e é frequentemente usado como um exemplo em tarefas de classificação de imagens e visão computacional. As informações contidas no conjunto de dados incluem características relacionadas à intensidade das cores, contrastes e outras medidas da imagem.



<ul>
<li>region-centroid-col: a coluna do pixel central da região.</li>
<li>region-centroid-row: a linha do pixel central da região.</li>
<li>region-pixel-count: o número de pixels em uma região = 9.</li>
<li>short-line-density-5: os resultados de um algoritmo de extração de linha que conta quantas linhas de comprimento 5 (qualquer orientação) com baixo contraste, menor ou igual a 5, passam pela região.</li>
<li>short-line-density-2: o mesmo que short-line-density-5, mas conta as linhas de alto contraste, maiores que 5.</li>
<li>vedge-mean: mede o contraste dos pixels adjacentes horizontalmente na região. Existem 6 valores, a média e o desvio padrão são fornecidos. Este atributo é usado como detector de borda vertical.</li>
<li>vedge-sd: desvio padrão de vedge-mean.</li>
<li>hedge-mean: mede o contraste dos pixels adjacentes verticalmente. Usado para detecção de linhas horizontais.</li>
<li>hedge-sd: desvio padrão de hedge-mean.</li>
<li>intensity-mean: a média sobre a região de (R + G + B) / 3.</li>
<li>rawred-mean: a média sobre a região do valor R.</li>
<li>rawblue-mean: a média sobre a região do valor B.</li>
<li>rawgreen-mean: a média sobre a região do valor G.</li>
<li>exred-mean: mede o excesso de vermelho: (2R - (G + B)).</li>
<li>exblue-mean: mede o excesso de azul: (2B - (G + R)).</li>
<li>exgreen-mean: mede o excesso de verde: (2G - (R + B)).</li>
<li>value-mean: transformação não linear em 3 dimensões do RGB.</li>
<li>saturation-mean: valor de saturação obtido a partir do valor RGB.</li>
<li>hue-mean: valor de matiz obtido a partir do valor RGB.</li>
</ul>
